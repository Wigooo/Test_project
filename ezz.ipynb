```python
ohe_col=["country","city","property_type"]
```

```python
df_encoded=pd.get_dummies(df,columns=ohe_col,prefix=ohe_col)
df_encoded
```

```python
df_encoded=df_encoded.drop(columns=drop_col,axis=1)
```

```python
df=pd.concat([df,df_encoded],axis=1)
df
```

```python
mapping = {
    'Unfurnished': 0,
    'Semi-Furnished': 1,
    'Fully-Furnished': 2
}
df['furnishing_status'] = df['furnishing_status'].map(mapping)
```

```python
df=df.drop(columns=ohe_col,axis=1)
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(columns="decision",axis=1)
y=df["decision"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.10,random_state=42)
```

```python
log_col = ["price" , "loan_amount" ,  "down_payment" ]
col2=["emi_to_income_ratio","customer_salary","monthly_expenses"]
```

```python
for col in log_col:
    x_train[col]=np.log(x_train[col]+1)
```

```python
for col in log_col:
    sns.histplot(x_train[col])
    plt.title(f"{col} hist after log transform")
    plt.show()
```

```python
x_train=x_train.loc[x_train["emi_to_income_ratio"]>0]
```

```python
x_train
```

```python
from sklearn.preprocessing import PowerTransformer # e3tmadd

# Yeo-Johnson handles zeros and skewness
yeojohnson = PowerTransformer(method='yeo-johnson')
x_train['emi_to_income_ratio_yeojohnson'] = yeojohnson.fit_transform(x_train[['emi_to_income_ratio']])
```

```python
x_train['customer_salary_sqrt'] = np.sqrt(x_train['customer_salary'])
```

```python
sns.histplot(x_train["emi_to_income_ratio_yeojohnson"])
```

```python
for col in col2:
    x_train[f"{col}_yeojohnson1"] = yeojohnson.fit_transform(x_train[[col]]).ravel()
    print(f"Applied Yeo-Johnson to {col}")
```

```python
sns.histplot(x_train["monthly_expenses_yeojohnson1"])
```
